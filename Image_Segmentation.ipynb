```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Load the image
image = cv2.imread('/content/up.jpg')
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Apply a threshold to create a binary image
# You may need to adjust the threshold value based on your image
threshold_value = 128
ret, binary_image = cv2.threshold(gray, threshold_value, 255, cv2.THRESH_BINARY)
```

```python
# Find contours in the binary image
contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Draw the contours on the original image
result_image = image.copy()
cv2.drawContours(result_image, contours, -1, (0, 255, 0), 2)  # -1 draws all contours
```

```python
# Display the original image with contours
cv2_imshow( result_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
